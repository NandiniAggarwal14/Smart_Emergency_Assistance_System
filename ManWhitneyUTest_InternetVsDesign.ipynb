```python
import pandas as pd
from scipy.stats import mannwhitneyu, rankdata
import matplotlib.pyplot as plt
```

```python

# Read the Excel file
file_path = r'C:\Users\prakr\Downloads\MannWhitneyUTest.xlsx'
df = pd.read_excel(file_path)

parameter = 'Consistency'

# Filter the data by group
internet = df[df['Group'] == 'Internet'][parameter].tolist()
designed = df[df['Group'] == 'Designed'][parameter].tolist()

print("Internet Ratings:", internet)
print("Designed Ratings:", designed)

# Combine for ranking
combined = internet + designed
ranks = rankdata(combined)

# Assign ranks back to groups
internet_ranks = ranks[:len(internet)]
designed_ranks = ranks[len(internet):]

# Manual U-stat calculation
n1, n2 = len(internet), len(designed)
R1, R2 = sum(internet_ranks), sum(designed_ranks)
U1 = R1 - n1 * (n1 + 1) / 2
U2 = R2 - n2 * (n2 + 1) / 2

print(f"\nSum of Ranks - Internet (R1): {R1}")
print(f"Sum of Ranks - Designed (R2): {R2}")
print(f"U1 (Internet): {U1}")
print(f"U2 (Designed): {U2}")
print(f"Smaller U used for test: {min(U1, U2)}")

# Mann–Whitney U Test (scipy)
stat, p = mannwhitneyu(internet, designed, alternative='two-sided')
print(f"\nscipy's Mann–Whitney U-statistic: {stat}")
print(f"scipy's P-value: {p}")

# Interpretation
alpha = 0.5
significant = p < alpha
```

```python
if significant:
    print("✅ Significant difference between Internet and Designed pop-ups.")
else:
    print("❌ No significant difference between the two groups.")


# -------------------- VISUALIZATION --------------------

# Plot average ratings for each group
groups = ['Internet', 'Designed']
avg_ratings = [sum(internet)/n1, sum(designed)/n2]

plt.figure(figsize=(6, 5))
bars = plt.bar(groups, avg_ratings, color=['#3498db', '#2ecc71'], edgecolor='black')

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.05, round(yval, 2), ha='center', va='bottom')

# Add significance annotation if applicable
if significant:
    ymax = max(avg_ratings)
    plt.text(0.5, ymax + 0.3, '* Significant (p < 0.05)', ha='center', color='red', fontsize=10)

# Graph labels
plt.title(f'{parameter} Rating Comparison\n(Internet vs Designed)')
plt.ylabel('Average Rating')
plt.ylim(0, max(avg_ratings) + 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
